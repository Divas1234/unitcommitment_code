```python
with open('.data/reformed_result.txt', 'r') as file:
    data = file.read()
print(data)
```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

# Convert the data to a numpy array
data_array = np.array([list(map(float, filter(None, row.split('\t')))) for row in data.split('\n') if row.strip()])

# Create a meshgrid for the surface plot
X, Y = np.meshgrid(range(data_array.shape[1]), range(data_array.shape[0]))
Z = data_array

# Plot the surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

plt.show()
```
